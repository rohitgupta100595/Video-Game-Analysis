```python
def find_missing_values(df):
    missing_count = df.isnull().sum()

    total_rows = len(df)
    missing_percentage = (missing_count / total_rows) * 100

    missing_data = pd.DataFrame({
        'Missing_Count': missing_count,
        'Missing_Percentage': missing_percentage.round(2)
    })

    missing_data = missing_data[missing_data['Missing_Count'] > 0].sort_values(
        by='Missing_Count', ascending=False
    )

    return missing_data
```

# *Checking unique values of Columns*

```python
def unique_col(df, max_show=10):
    for col in df.columns:
        uniques = df[col].unique()
        n_uniques = len(uniques)
        
        print(f"\nColumn: {col}")
        print(f"Unique count: {n_uniques}")
        if n_uniques > max_show:
            print(f"Sample unique values: {uniques[:max_show]} ...")
        else:
            print(f"Unique values: {uniques}")
```

```python
unique_col(games)
```

```python
unique_col(vgsales)
```

# *Converting release date column in games to date*

```python
games['Release Date'].unique()
```

```python
games['Release Date'] = pd.to_datetime(games['Release Date'],errors='coerce')
```

```python
games['Release Date'].isna().sum()
```

```python
# The NA in Date Column is TBD, hence we will remove it.
rows_to_drop_games = games[games['Release Date'].isna()].index
```

```python
games.drop(rows_to_drop_games,inplace=True)
```

```python
# Fill NA in Rating with 0 and Team with Unknown
games['Rating'] = games['Rating'].fillna(0)
games['Team'] = games['Team'].fillna('Unknown')
```

```python
def convert_to_number(value):
    if isinstance(value, str):
        value = value.strip().upper()
        
        if value.endswith("K"):
            return float(value[:-1]) * 1000
        elif value.endswith("M"):
            return float(value[:-1]) * 1000000
        elif value.endswith("B"):
            return float(value[:-1]) * 1000000000
        else:
            try:
                return float(value)  # if it's just a number string
            except ValueError:
                return None  # if not convertible
    return value  # return as is if already numeric
```

```python
# Converting object type ['Times Listed','Number of Reviews','Plays','Playing','Backlogs','Wishlist'] to float type
games['Times Listed'] = games['Times Listed'].apply(convert_to_number)
games['Number of Reviews'] = games['Number of Reviews'].apply(convert_to_number)
games['Plays'] = games['Plays'].apply(convert_to_number)
games['Playing'] = games['Playing'].apply(convert_to_number)
games['Backlogs'] = games['Backlogs'].apply(convert_to_number)
games['Wishlist'] = games['Wishlist'].apply(convert_to_number)
```

```python
def safe_literal_eval(x):
    if isinstance(x, str):
        try:
            # Try to convert the string to a list/tuple
            evaluated_list = ast.literal_eval(x)
            
            # Now, check if the result is actually a list/tuple before joining
            if isinstance(evaluated_list, (list, tuple)):
                return ", ".join(evaluated_list)
            else:
                # Handle cases where the string evaluates to a single item (e.g., "'Team A'")
                return str(evaluated_list)
                
        except (ValueError, TypeError, SyntaxError):
            # If parsing fails, just return the original string
            return x
    return x

# Apply the function
games['Team'] = games['Team'].apply(safe_literal_eval)
games['Genres'] = games['Genres'].apply(safe_literal_eval)
```

```python
# Removing Summary and Reviews column as we dont need them
games.drop(columns = ['Summary','Reviews'],axis=1,inplace=True)
```
